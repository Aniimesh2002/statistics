Question 1: What is hypothesis testing in statistics?

answer: Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data.
It helps us determine whether there is enough evidence in the sample data to support or reject a claim (hypothesis) about the population.



Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

answer:

# **1. Null Hypothesis (H₀):**

It is the default assumption or status quo.
States that there is no effect, no difference, or no relationship in the population.
It is what we assume to be true until evidence proves otherwise.

Example:

A medicine has no effect on blood pressure.

The average exam score = 70.

A coin is fair (probability of heads = 0.5).

# **2. Alternative Hypothesis (H₁ / Ha):**

It is the opposite of H₀.

States that there is an effect, difference, or relationship.

This is what the researcher is usually trying to prove or provide evidence for.

Example :

A medicine does affect blood pressure.

The average exam score ≠ 70.

A coin is not fair (probability of heads ≠ 0.5).


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

answer:

The significance level (α) is the threshold probability we set before testing that helps decide whether to reject the null hypothesis (H₀).

It represents the maximum risk of making a wrong decision (rejecting H₀ when it is actually true).
This wrong decision is called a Type I error.

Question 4: What are Type I and Type II errors? Give examples of each.

answer:

Type I Error (False Positive)

Occurs when we reject the null hypothesis (H₀) even though it is true.

In other words, we conclude there is an effect/difference when actually there isn’t.

Probability of making this error = α (significance level).

Example 1 (Medicine):

H₀: A new drug has no effect.

If we reject H₀ (say it works) but in reality, it doesn’t → Type I error.


Type II Error (False Negative)>-

Occurs when we fail to reject the null hypothesis (H₀) even though it is false.

In other words, we miss detecting an actual effect or difference.

Probability of making this error = β.

Example 1 (Medicine):

H₀: A new drug has no effect.

If we fail to reject H₀ (say it doesn’t work) but in reality, it does → Type II error



Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

answer:

Both Z-test and T-test are statistical hypothesis tests used to compare sample data with population parameters or between groups. The main difference comes from sample size and whether population standard deviation is known.


Z-Test

Used when:

The population standard deviation (σ) is known.

The sample size is large (n > 30) → Central Limit Theorem applies.

Distribution used: Standard Normal distribution (Z-distribution).



T-Test

Used when:

The population standard deviation (σ) is unknown.

The sample size is small (n ≤ 30).

Distribution used: Student’s t-distribution, which has heavier tails than normal (to account for extra uncertainty with small samples).


```python
#  Question 6: Write a Python program to generate a binomial distribution with n=10 and
#  p=0.5, then plot its histogram.



import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(-0.5, n+1.5, 1), density=True, edgecolor='black')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

```

```python
"""

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


"""



import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)   # sample standard deviation
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample Mean = {sample_mean:.3f}")
print(f"Sample Std Dev = {sample_std:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.5f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is significant evidence that the mean is different from 50.")
else:
    print("Fail to Reject H0: No significant evidence that the mean differs from 50.")



'''
Interpretation

The sample mean ≈ 50.03, very close to the hypothesized mean (50).
Z = 0.466 indicates the sample mean is less than 0.5 standard errors away from 50.
p-value ≈ 0.64 > 0.05, so we fail to reject H₀.

Conclusion: There is no significant evidence that the true mean is different from 50.
'''
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # population mean and std dev
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Step 3: 95% Confidence Interval
confidence_level = 0.95
z_value = stats.norm.ppf(1 - (1-confidence_level)/2)  # 1.96 approx for 95%
ci_lower = sample_mean - z_value * se
ci_upper = sample_mean + z_value * se

print(f"Sample Mean = {sample_mean:.3f}")
print(f"95% Confidence Interval = ({ci_lower:.3f}, {ci_upper:.3f})")

# Step 4: Plot histogram of data
plt.hist(data, bins=15, density=True, alpha=0.6, edgecolor="black")
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='blue', linestyle=':', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle=':', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores represent
#in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    # Compute mean and std
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    # Calculate z-scores
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
data = np.array([10, 12, 9, 15, 20, 18, 22, 17, 13, 14])

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print results
print("Original Data:", data)
print("Z-scores:", np.round(z_scores, 2))

# Plot histogram of Z-scores
plt.hist(z_scores, bins=6, color="skyblue", edgecolor="black", density=True)
plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency (density)")
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.show()

```
